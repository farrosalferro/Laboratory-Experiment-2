```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing file
col_names = ['Time', 'None', 'G', 'B', 'R']
df_1 = pd.read_csv('1.csv', skiprows=4, names=col_names) * 1000
df_40 = pd.read_csv('40.csv', skiprows=4, names=col_names) * 1000
df_16 = pd.read_csv('16.csv', skiprows=4, names=col_names) * 1000
df_18 = pd.read_csv('18.csv', skiprows=4, names=col_names) * 1000
df_20 = pd.read_csv('20.csv', skiprows=4, names=col_names) * 1000
print(df_1.head())
print(df_1.shape)
```

```python
sns.set()
sns.set_style('darkgrid', {'grid.color': '.55'})
sns.set_context('paper')
sns.set_palette('husl')
```

```python
df_16_1 = df_16[(-21.375 <= df_16['Time']) & (df_16['Time'] <= -20.575)]
df_18_1 = df_18[(-18.375 <= df_18['Time']) & (df_18['Time'] <= -17.225)]
df_20_1 = df_20[(-17.100 <= df_20['Time']) & (df_20['Time'] <= -15.475)]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,8), sharey=True)
ax[0].plot(df_16_1['Time'], df_16_1['G'], c='g')
ax[0].plot(df_16_1['Time'], df_16_1['B'], c='b')
ax[0].plot(df_16_1['Time'], df_16_1['R'], c='r')

ax[1].plot(df_18_1['Time'], df_18_1['G'], c='g')
ax[1].plot(df_18_1['Time'], df_18_1['B'], c='b')
ax[1].plot(df_18_1['Time'], df_18_1['R'], c='r')

ax[2].plot(df_20_1['Time'], df_20_1['G'], c='g')
ax[2].plot(df_20_1['Time'], df_20_1['B'], c='b')
ax[2].plot(df_20_1['Time'], df_20_1['R'], c='r')

ax[0].set_ylim(-100, 1800)
ax[1].set_ylim(-100, 1800)
ax[2].set_ylim(-100, 1800)
fig.supxlabel('Time (ms)')
fig.supylabel('Voltage (mV)')
plt.savefig('../results/16 vs 18 vs 20.png', bbox_inches='tight')
plt.show()
```

```python
results = [1, 2, 3, 4, 5, 10, 15, 16, 18, 20, 21, 25, 27, 30, 35, 36, 38, 40]
col_names = ['Time', 'None', 'G', 'B', 'R']
for result in results:
    file_name = str(result) + '.csv'
    file_output = '../results/' + str(result) + '.png'
    df = pd.read_csv(file_name, skiprows=4, names=col_names) * 1000
    
    fig, ax = plt.subplots(figsize=(12,8))
    ax.plot(df['Time'], df['G'], c='g', label='Green')
    ax.plot(df['Time'], df['B'], c='b', label='Blue')
    ax.plot(df['Time'], df['R'], c='r', label='Red')
    ax.plot(df['Time'], df['None'], c='y', label='No Filter')
    
    ax.set_xlabel('Time (ms)')
    ax.set_ylabel('Voltage (mV)')
    ax.set_ylim(-200, 5000)
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
            ncol=4, fancybox=True, shadow=True)
    plt.savefig(file_output, bbox_inches='tight')
```
