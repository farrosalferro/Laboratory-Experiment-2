```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

column_names = ['Time', 'x_nogyro', 'y_nogyro', 'z_nogyro']
df_nogyro = pd.read_csv('NoGyro-IMU.csv', names=column_names)
print(df_nogyro.head())
```

```python
sns.set()
sns.set_style('darkgrid')
sns.set_palette('husl')
```

```python
fig_nogyro = plt.figure(figsize=(16,16))
ax_nogyro = fig_nogyro.add_subplot(projection='3d')

ax_nogyro.plot(df_nogyro['x_nogyro'], df_nogyro['y_nogyro'], df_nogyro['z_nogyro'], linewidth=3)
ax_nogyro.set_xlabel('X Label')
ax_nogyro.set_ylabel('Y Label')
ax_nogyro.set_zlabel('Z Label')
plt.show()
```
