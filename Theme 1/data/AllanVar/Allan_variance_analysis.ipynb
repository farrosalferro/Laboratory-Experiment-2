```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['tau', 'avar_ax', 'avar_ay', 'avar_az']
df = pd.read_csv('AllanVar_Acc.csv', names=column_names)
print(df.head())
```

```python
sns.set()
sns.set_style('darkgrid')
sns.set_palette('husl')

fig, axes = plt.subplots(figsize=(8, 8))

axes.scatter(df['tau'], df['avar_ax'], edgecolors='r', facecolors='none', s=200)
axes.plot(df['tau'], df['avar_ax'], color='r', label='X')
axes.scatter(df['tau'], df['avar_ay'], edgecolors='g', facecolors='none', s=200)
axes.plot(df['tau'], df['avar_ay'], color='g', label='Y')
axes.scatter(df['tau'], df['avar_az'], edgecolors='b', facecolors='none', s=200)
axes.plot(df['tau'], df['avar_az'], color='b', label='Z')
axes.set_yscale('log')
axes.set_xscale('log')
axes.set_xlabel('Tau, s')
axes.set_ylabel('Sigma, m/s^2')

plt.legend()
plt.show()
```

```python
column_names_g = ['tau', 'avar_gx', 'avar_gy', 'avar_gz']
df_g = pd.read_csv('AllanVar_Gyro.csv', names=column_names_g)
print(df_g.head())
```

```python
fig, axes = plt.subplots(figsize=(8, 8))

axes.scatter(df_g['tau'], df_g['avar_gx'], edgecolors='r', facecolors='none', s=200)
axes.plot(df_g['tau'], df_g['avar_gx'], color='r', label='X')
axes.scatter(df_g['tau'], df_g['avar_gy'], edgecolors='g', facecolors='none', s=200)
axes.plot(df_g['tau'], df_g['avar_gy'], color='g', label='Y')
axes.scatter(df_g['tau'], df_g['avar_gz'], edgecolors='b', facecolors='none', s=200)
axes.plot(df_g['tau'], df_g['avar_gz'], color='b', label='Z')
axes.set_yscale('log')
axes.set_xscale('log')
axes.set_xlabel('Tau, s')
axes.set_ylabel('Sigma, dps')

plt.legend()
plt.show()
```

```python
ratio_x = df['avar_ax'] / df_g['avar_gx']
ratio_y = df['avar_ay'] / df_g['avar_gy']
ratio_z = df['avar_az'] / df_g['avar_gz']
print(f'ratio x: \n{ratio_x.mean()}')
print(f'ratio y: \n{ratio_y.mean()}')
print(f'ratio z: \n{ratio_z.mean()}')

```
